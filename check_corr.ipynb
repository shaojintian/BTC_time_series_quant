```python
import pandas_datareader.data as web
import datetime
import pandas as pd

# 定义起始和结束日期
start_date = datetime.datetime(2020, 1, 1)  # 起始日期，可以根据需要修改
end_date = datetime.datetime.now()  # 结束日期，将使用当前日期

# 纳斯达克100指数的代码，这里以QQQ为例，它是纳斯达克100 ETF的跟踪基金
nasdaq_code = 'QQQ'

# 使用pandas-datareader从Yahoo Finance下载数据
data = pd.DataFrame()
try:
    data = web.DataReader(nasdaq_code, 'yahoo', start_date, end_date)
    print(data.head())  # 打印前几行数据
except Exception as e:
    print(f"An error occurred: {e}")

# 保存数据到CSV文件
data.to_csv('nasdaq100_data.csv')

```
